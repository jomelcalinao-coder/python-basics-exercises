### 1. Print characters present at an even index number


#### Write a Python code to accept a string from the user and display characters present at an even index number.

For example, str = "PYnative". so your code should display ‘P’, ‘n’, ‘t’, ‘v’.

Expected Output:

Orginal String is  PYnative
Printing only even index chars
P
n
t
v

```python
# User types a word. In this instance, the word is "PYnative".
text = input("Enter a string: ")

#Print the required text
print("Orginal String is", text)
print("Printing only even index chars")

#Display the word at an even index number per line
for char in text[::2]:
    print(char)
```

### 2. Check if the first and last numbers of a list are the same

Write a code to return True if the list’s first and last numbers are the same. If the numbers are different, return False.

Given:

numbers_x = [10, 20, 30, 40, 10]
Output True

numbers_y = [75, 65, 35, 75, 30]
Output False



```python
def check_first_last_same(numbers):
    return numbers[0] == numbers[-1]


numbers_x = [10, 20, 30, 40, 10]
numbers_y = [75, 65, 35, 75, 30]

print(check_first_last_same(numbers_x))  # True
print(check_first_last_same(numbers_y))  # False
```

```python
def check_first_last_same(numbers):
    if numbers[0] == numbers[-1]:
        return True
    else:
        return False


numbers_x = [10, 20, 30, 40, 10]
numbers_y = [75, 65, 35, 75, 30]

print(check_first_last_same(numbers_x))
print(check_first_last_same(numbers_y))
```
