### 1. Print characters present at an even index number


Write a Python code to accept a string from the user and display characters present at an even index number.

For example, str = "PYnative". so your code should display ‘P’, ‘n’, ‘t’, ‘v’.

Expected Output:

Orginal String is  PYnative
Printing only even index chars
P
n
t
v

```python
# User types a word. In this instance, the word is "PYnative".
text = input("Enter a string: ")

#Print the required text
print("Orginal String is", text)
print("Printing only even index chars")

#Display the word at an even index number per line
for char in text[::2]:
    print(char)
```
